Modeling simple AND, OR, and XOR functions using a perceptron

# AND 
F(x) =w1*x1 + w2*x2 + b = 0
1.   w1 = w2 = 1
2.   b = -3/2

# OR
F(x) = w1*x1 + w2*x2 + b = 0
1.   W1 = W2 = 1
2.   B = -1/2

XOR = OR - AND

```
import numpy as np
import matplotlib.pyplot as plt

def perceptron(inputs, weights, bias):
  pred = []
  for x in inputs:
    net = np.dot(weights, x) + bias
    if net >= 0:
      y = +1
    else:
      y = 0
    pred.append(y)
  return pred  
```

```
# AND
w = np.array([1,1])
b = -3/2

# load my data
x = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
t = np.array([0,0,0,1])

x1 = np.arange(-0.1,1.5,0.01) # you can use also np.linespace(0,1,100)
x2 = -w[0]/w[1] * x1 - b/w[1]

plt.scatter(x[:,0],x[:,1],c=t, marker="s", s=100)
plt.plot(x1,x2)
plt.grid()
plt.show()
```

```
# new figure
x1 = np.linspace(-0.1,1.1,25)
x2 = np.linspace(-0.1,1.1,25)
xg, yg = np.meshgrid(x1, x2)

zg = perceptron(np.hstack((xg.reshape(-1,1), yg.reshape(-1,1))), w, b)

plt.scatter(xg, yg, c=zg)
colors = np.array(["red","red","red","black"])
plt.scatter(x[:,0], x[:,1], c=colors, marker="s", s=100)
```

```
# OR
w = np.array([1,1])
b = -1/2

# load my data
x = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
t = np.array([0,1,1,1])

x1 = np.arange(0,1,0.01) # you can use also np.linespace(0,1,100)
x2 = -w[0]/w[1] * x1 - b/w[1]

plt.scatter(x[:,0],x[:,1],c=t, marker="s", s=100)
plt.plot(x1,x2)
plt.grid()
plt.show()

# new figure
plt.figure()
x1 = np.linspace(-0.1,1.1,100)
x2 = np.linspace(-0.1,1.1,100)
xg, yg = np.meshgrid(x1, x2)

zg = perceptron(np.hstack((xg.reshape(-1,1), yg.reshape(-1,1))), w, b)

plt.scatter(xg, yg, c=zg)
colors = np.array(["red","black","black","black"])
plt.scatter(x[:,0], x[:,1], c=colors, marker="s", s=100)
```

#XOR
It requires two decision lines and 3 perceptrons

```
import numpy as np
import matplotlib.pyplot as plt

def hardlim(net):
  y = []
  for i in net:
    if i >= 0:
      y.append(1)
    else:
      y.append(0)
  return y

def fnn(inputs):
  y = []
  for x in inputs:
    h = hardlim(np.dot(w1,x)+b1)        # Hidden layer
    y.append(hardlim(np.dot(w2,h)+b2))  # Output layer
  return y

# weigts of hidden layer
w1 = np.array([[1,1],[1,1]])
b1 = np.array([-3/2, -1/2])
# weights of output layer
w2 = np.array([-1,1])
b2 = np.array([-1/2])

# load my data
x = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
t = np.array([0,1,1,0])

# feedforward neural network
y = fnn(x)
print('Targets:\n',t)
print('Predictions:\n',y)

# new figure
plt.figure()
x1 = np.linspace(-0.1,1.1,100)
x2 = np.linspace(-0.1,1.1,100)
xg, yg = np.meshgrid(x1, x2)

zg = fnn(np.hstack((xg.reshape(-1,1), yg.reshape(-1,1))))

plt.scatter(xg, yg, c=zg)
colors = np.array(["red","black","black","red"])
plt.scatter(x[:,0], x[:,1], c=colors, marker="s", s=100)
```
