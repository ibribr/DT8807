```python
from transformers import pipeline

# Define the zero-shot classification pipeline using GPT-2
classifier = pipeline("zero-shot-classification", model="EleutherAI/gpt-neo-1.3B")
```

```python
# Define the possible labels for the animals
labels = ["mammal", "reptile", "bird"]

# Define the text descriptions of the animals to classify
animals = [
    "A furry creature that nurses its young with milk",
    "A cold-blooded animal that lays eggs and has scales",
    "A four-legged animal that chews cud and has hooves",
    "A slimy animal that lives in water and has a shell",
    "A creature with feathers that can fly",
]

# Perform zero-shot classification on each animal description
for animal in animals:
    result = classifier(animal, labels)
    print(f"{animal}: {result['labels'][0]}")
```
